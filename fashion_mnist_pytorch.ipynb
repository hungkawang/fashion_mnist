```python
import torch
from torch import nn

from torchvision import datasets
from torch.utils.data import DataLoader
from torchvision.transforms import ToTensor
```

```python
train_data = datasets.FashionMNIST(
    root = "data", 
    train = True,
    download = True, 
    transform = ToTensor()
)
```

```python
train_dataloader = DataLoader(dataset = train_data, batch_size = 32)
```

```python
test_dataloader = DataLoader(dataset = test_data, batch_size = 32)
```

```python
import torch.nn as nn
import torch.nn.functional as F


class Net(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = nn.Conv2d(1, 6, 5)
        self.pool = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(6, 16, 5)
        self.fc1 = nn.Linear(16 * 4 * 4, 120)
        self.fc2 = nn.Linear(120, 84)
        self.fc3 = nn.Linear(84, 10)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = torch.flatten(x, 1) # flatten all dimensions except batch
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x


net = Net()
```

```python
loss_f = nn.CrossEntropyLoss()
```

```python
optimizer = torch.optim.SGD(params = net.parameters(), lr = 1e-3)
```

```python
def training(dataloader, model, loss_f, optimizer):
    size = len(dataloader.dataset)
    
    model.train()
    
    for batch, (x, y) in enumerate(dataloader):
        pred = model(x)
        loss = loss_f(pred, y)
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        if batch % 100 == 0:
            loss = loss.item()
            current = batch * len(x)
            print(f"loss: {loss:>5f} at {current:>5d}/{size:>5d}")
```

```python
def testing(dataloader, model, loss_f):
    size = len(dataloader.dataset)
    num_batch = len(dataloader)
    
    model.eval()
    
    test_loss, correct = 0, 0
    
    with torch.no_grad():
        for x, y in dataloader:
            pred = model(x)
            test_loss += loss_f(pred, y).item()
            correct += (pred.argmax(1) == y).type(torch.float).sum().item()
            
    test_loss /= num_batch
    correct /= size
    
    print(f"\n Test Evaluation: \n Accuracy: {(100*correct):>5f}%, Avg loss: {test_loss:>5f} \n")
```

```python
epochs = 10

for t in range(epochs):
    print(f"\n\n  Epoch {t+1}  --------------------\n")
    
    training(train_dataloader, net, loss_f, optimizer)
    
    testing(test_dataloader, net, loss_f)
```
